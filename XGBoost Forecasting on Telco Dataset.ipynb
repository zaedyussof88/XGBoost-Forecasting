## Importing Dataset

```python
df = pd.read_csv("Telco Data.csv")
```

## Cleaning Dataset

### Finding attributes with missing cells

```python
df.isna().sum()   
```

### Further information

```python
df.info()
```

### Drop columns with missing cells

```python
#missing value terlalu banyak and impossible to impute
df = df.drop(['logtoll', 'logequi', 'logcard', 'logwire'], axis=1)
```

```python
df.info()
```

### Create Dataframe of the targeted column

```python
y = df['unsubscribe']
df2 = df.drop(['unsubscribe'], axis=1)
```

### Displaying number of unique values per column

```python
df2.nunique()
```

### Drop columns with less unique values

```python
df3 = df2.drop(['ebill','confer', 'forward', 'callwait', 'callid', 'internet', 'pager', 'voice', 'multline','wireless', 'callcard', 'equip', 'tollfree', 'gender', 'retire', 'marital'], axis=1)
```

# Create Correlation Graph

```python
df3.corrwith(df['unsubscribe']).sort_values(ascending = False).plot.bar(figsize = (30, 15), title = 'Correlation with Churn', fontsize = 20, rot = 45, grid = True)
```

## Feature Importance

```python
# Another way to create XGBoost same like down coding
clf = xgb.XGBClassifier(max_depth = 7, n_estimators = 100, colsample_bytree = 0.8, subsample = 0.8, nthread = 10, learning_rate = 0.05)
clf.fit(df3,y)
```

```python
# Creating an XGBoost classifier with specified parameters
clf = xgb.XGBClassifier(
    max_depth=7,
    n_estimators=100,
    colsample_bytree=0.8,
    subsample=0.8,
    nthread=10,
    learning_rate=0.05
)

# Training the classifier on your data
clf.fit(df3, y)
```

## Plot important features

```python
fig, ax = plt.subplots(figsize=(12,18))
xgb.plot_importance(clf, max_num_features=50, height=0.8, ax=ax)
plt.show()
```

## Data Splitting with Stratified Sampling

```python
xtrain, xvalid, ytrain, yvalid = train_test_split(df2, y, stratify=y, random_state=1, test_size=0.2, shuffle=True)
```

## XGBoost Classifier Training and Evaluation

```python
clf = xgb.XGBClassifier(max_depth=7, n_estimators=100, colsample_bytree=0.8, subsample=0.8, learning_rate=0.05)
clf.fit(xtrain, ytrain)
predictions = clf.predict(xvalid)
print("accuracy score =", accuracy_score(yvalid, predictions))
```
